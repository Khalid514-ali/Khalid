```python
with open('app.py','w') as f:
  f.write("""import streamlit as st
from PyPDF2 import PdfReader
import torch
import soundfile as sf
import os
from TTS.api import TTS

st.set_page_config(page_title="PDF to Audiobook", layout="centered")

st.title("PDF → Audiobook (Voice Cloning)")
st.write("Upload a PDF and a voice sample to create an audiobook in the same voice")

# Create temp folder
os.makedirs("temp", exist_ok=True)

# Upload PDF
pdf_file = st.file_uploader("Upload PDF", type=["pdf"])

# Upload voice sample
voice_file = st.file_uploader("Upload Voice Sample (5–15 sec)", type=["wav", "mp3"])

if pdf_file and voice_file:
    # Save voice sample
    sample_path = "temp/sample.wav"
    with open(sample_path, "wb") as f:
        f.write(voice_file.read())

    # Extract text from PDF
    reader = PdfReader(pdf_file)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"

    st.subheader("Extracted Text Preview")
    st.text_area("PDF Text", text[:2000], height=200)

    if st.button("Generate Audiobook"):
        with st.spinner("Generating audiobook... this may take time ⏳"):
            device = "cuda" if torch.cuda.is_available() else "cpu"

            tts = TTS(
                model_name="tts_models/multilingual/multi-dataset/xtts_v2",
                progress_bar=False
            ).to(device)

            output_path = "temp/audiobook.wav"

            tts.tts_to_file(
                text=text,
                speaker_wav=sample_path,
                language="en",
                file_path=output_path
            )

        st.success("Audiobook generated!")

        # Audio player
        audio_bytes = open(output_path, "rb").read()
        st.audio(audio_bytes, format="audio/wav")

        # Download button
        st.download_button(
            label=" Download Audiobook",
            data=audio_bytes,
            file_name="audiobook.wav",
            mime="audio/wav"
        ))""")
```
